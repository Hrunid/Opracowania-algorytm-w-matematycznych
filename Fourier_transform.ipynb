# Muzyka, matemaktyka i informatyka
## 1. **Muzyka**
### **Struktura stroju równomiernie temperowanego - czyli dlaczego mamy 12 dźwięków w oktawie?**
#### **Czym jest oktawa?**
Oktawa jest interwałem, odstępem między dwoma dźwiękami. Z fizycznego punktu widzenia oktawą jest różnica podwojonej i bazowej częstotliwości dźwięku, więc dla dźwięku A4 = $440Hz$ oktawą wyżej będzie dźwięk o częstotliwości $880Hz$, czyli A5, a oktawę niżej mamy A3 = $220Hz$. Ta proporcja nie jest przypadkowa. Nasze ucho naturalnie słyszy oktawę jako "ten sam dźwięk", ale wyżej, o innej barwie. 
#### **Składowe harmoniczne**
Jednym z powodów takiego wrażenia jest natura dźwięków produkowanych przez klasyczne instrumenty: otóż oprócz częstotliwości podstawowej każdy dźwięk składa się również z jej wielokrotności i noszą one nazwę **składowych harmonicznych** lub **widma harmonicznego**. Tak więc oprócz częstotliwości bazowej $f_0$ w skład dźwięku wchodzą również $2(f_0), 3(f_0), ..., n(f_0)$. Przy czym poszczególne składowe nie muszą i prawie nigdy nie stanowią równych części. Tutaj możemy zauważyć powiązanie z oktawą - jej pierwszą składową harmoniczą jest częstotliwość o podwójnej wartości względem bazowej.
#### **Barwa**
Barwę możemy określić lub rozpoznać właśnie ze względu na proporcje składowych harmonicznych. Ten fakt będzie kluczowy przy cyfrowej analizie sygnału, przy problemach np. rozpoznawania instrumentów czy indentyfikowania ludzi po głosie. O niektórych zastosowaniach w dalszej częsci oprawcowania.
#### **Kontekst historyczny**
Wyżej wspominam konkretne wartości liczbowe, skład harmoniczny, ale przecież ludzie starożytni dali początek systemowi, zanim jestcze dostępna była aparatura umożliwiająca badanie tak precyzyjnych właściwości fal dźwiękowych. Ich odkrycia wizięły się głownie z wrażeń słychowych oraz prostego, ale jednak genialnego ekperymentu: otóż przez podzielenie struny na pół uzyskali uzyskali efekt opisany powyżej, czyli wrażenie dźwięku bardzo zbliżonego, ale jednak wyższego.

Kolejnym niezwykle ważnym interwałem była **kwinta**, która powstała przez dalsze podzielenie struny na pół(oktawy, nie całości) i znowu wrażenia słuchowego jako bardzo przyejmengo dla ucha. Jeśli oktawa dawała proporcje $2:1$, to po kolejnym podzieleniu dostajemy proporcję kwinty: $3:2$. Te dwie proporcje dały początek obencenu podziałowi na oktawy i jej dwanaście dźwięków.
#### **Rola matematyki**
No dobrze, ale mamy dopiero dwa interwały. Dość daleko do 12. Mając te dwa naturalne interwały można współczesny wachlarz dźwięków uzyskać dzięki matematyce. Jeśli weźmiemy dowolny dźwięk o częstotliwości $f_0$, to przechodzenie oktawami w góre i dół można opisać ciągiem:
$$
a_n = f_0 \cdot 2^n, n \in \mathbb{Z}
$$
A dla kwint mamy ciąg:
$$
b_m = f_0 \cdot \left(\tfrac{3}{2}\right)^m, m \in \mathbb{Z}
$$
Na poziomie praktyki muzycznej widać to jako tzw. koło kwint: zaczynając od C i idąc 12 razy w górę kwintą, przechodzimy przez wszystkie dźwięki systemu aż wracamy – enharmonicznie – do C:  
$$
C \rightarrow G \rightarrow D \rightarrow A \rightarrow E \rightarrow H \rightarrow F\sharp \rightarrow C\sharp \rightarrow G\sharp \rightarrow D\sharp \rightarrow A\sharp \rightarrow F \rightarrow C.
$$ 
Wróćmy jednak do matematyki, bo pojawia się pewna nieścisłość. Nie bez powodu użyłem stwierdzenia "bardzo blisko", ponieważ po przejściu 12 kwint oraz 7 oktaw otrzymujemy następującje wartości ciągów z pominięciem $f_0$: 
$$
a_7 = 2^7
b_{12} = \left(\tfrac{3}{2}\right)^{12}
$$
Co daje:
$$
128 \approx 129,7463
$$
Czyli jest pewnien problem; Nie mamy równości, co w praktyce oznacza stopniową rozbieżność stroju, ponieważ budując z idealnych kwint dostajemy zwiększający się błąd o $\frac{129,7463}{128} \approx 1,0136 $ W systemie równomiernie temperowanym ten błąd rozkłada się po wszystkich interwałach: zamiast idealnego $\frac{3}{2}$ przyjmuje się **kwintę równomiernie temperowaną** o stosunku  
$$
2^{7/12} \approx 1,4983
$$
trochę skorygowaną i niedokładną w stosunku do 1,5, ale za to taką, że 12 takich kwint daje dokładnie 7 oktaw. Bacha interesowało właśnie takie uśrednienie błędu: drobne rozstrojenie mniejszych interwałów w zamian za to, że strój nie rozjeżdża się stopniowo przy kolejnych modulacjach, a wszystkie tonacje stają się używalne – co pokazał praktycznie w „Das wohltemperierte Klavier".

## 2. **Matematyka**
### 2.1 **Czym jest dźwięk?**
Dźwięk jest falą rozchodzącą się ośrodku, która powoduje zmiany ciścnienia powietrza. Dla wrażeń słuchowych kluczowe są trzy właściwości:

- częstotliwość 
- amplituda - głośność
- skład harmoniczny widma

Matematycznie dźwięk możemy określić jako funkcję(zazwyczaj ciągłą) opisującą zmiany ciśnienia powietrza $f(t)$, której dziedziną jest czas. Dla dźwięku o konkretnej wysokości przebieg $f(t)$ można traktować jako funkcję okresową, ponieważ wysokości dźwięków opisujemy za pomocą częstotliwości.
Funkcje okresowe spełniające warunki Dirichleta da się przedstawić w potaci szeregu Fouriera:

$$
f(x) = \frac{a_0}{2} + \sum_{n=1}^{\infty} \left( a_n \cos(nx) + b_n \sin(nx) \right)
$$
Istotne jest tutaj uściślenie, że funckja dźwięku na ogół nie jest okresowa, dlatego rozpatrujemy ją jako lokalnie okresową tj. w pewnym przedziale jej dziedziny. W praktyce oznacza to podział sygnału na krótkie okna czasowe i rozwijanie ich w szereg Fouriera.
### **2.2 Analiza składu harmonicznego przy pomocy szeregu Fouriera**
Rozważmy teraz konkrenty przykład. Powiedzmy, że mamy dźwięk o częstotliowści podstawowej $f_0 = 261Hz$, co daje nam dźwięk C czwartej oktawy. Wzór funckji $T$-okresowej rozwiniętej w szereg Fouriera wygląda następująco:
$$
f(x) = \frac{A_0}{2} + \sum_{n=1}^{\infty}\left(A_n cos\left(\frac{2\pi x}{T}\right)\right) + \left(B_n sin\left(\frac{2\pi x}{T}\right)\right)
$$
Jeśli nasz dźwięk ma częstotliwość $f_0$, to okres $T = \frac{1}{f_0}$. Dlatego w przedziale czasowym $[a;b]$, kiedy występuje dany dźwięk funkcja jest $T$-okresowa. Stąd dla $f_0$ otrzymujemy szereg:

$$
f(t) = \frac{A_0}{2} + \sum_{n=1}^{\infty}\left(A_n cos\left(2\pi t f_0\right)\right) + \left(B_n sin\left(2\pi t f_0\right)\right)
$$
Zauważmy jednak dwie rzeczy. Dla prostych przypadków wykres dźwięku możemy reprezentować przy pomocy jednynie sinusa, więc zakładamy, że $A_n = 0$. Dodatkowo wyrazami pozostałego $B_n$ są właśnie proporcje składowych harmonicznych ($n$-ty współczynnik to ilość widma o częstotliwość $n$ ), a znając częstotliwość bazową, znamy też wartości kolejnych składowych harmonicznych, a współczynniki szeregu Fouriera dokładnie określają, ile której ze składowych zawiera dźwięk. Ponieżej aproksymacja wykresów dźwiękowych różnych instrumentów a obok wykres proporcji składowych harmonicznych. O tym, jak z dyskretnego sygnału dźwiękowego uzyskać widmo w kolejnej części.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry
fs = 44100          # Częstotliwość próbkowania
f0 = 260.0          # C4
duration = 0.02     # 20 ms
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# Przybliżenie realnych współczynnikiów
flute_coeffs  = np.array([1.0, 0.3, 0.15, 0.05, 0.03, 0.02, 0.01, 0.005])
piano_coeffs  = np.array([1.0, 0.7, 0.5,  0.3,  0.2,  0.15, 0.1,  0.08 ])
violin_coeffs = np.array([1.0, 0.9, 0.8,  0.6,  0.4,  0.3,  0.25, 0.2 ])

def synth_from_fourier(f0, t, coeffs):
    signal = np.zeros_like(t)
    for n, B in enumerate(coeffs, start=1):
        signal += B * np.sin(2 * np.pi * n * f0 * t)
    signal /= np.max(np.abs(signal))
    return signal

sinus   = np.sin(2 * np.pi * f0 * t)
sinus  /= np.max(np.abs(sinus))

flute  = synth_from_fourier(f0, t, flute_coeffs)
piano  = synth_from_fourier(f0, t, piano_coeffs)
violin = synth_from_fourier(f0, t, violin_coeffs)

signals = [
    ("Czysty ton (sinus)", sinus,   np.array([1.0])),
    ("Flet",               flute,  flute_coeffs),
    ("Fortepian",          piano,  piano_coeffs),
    ("Skrzypce",           violin, violin_coeffs),
]

plt.figure(figsize=(12, 10))

for i, (name, sig, coeffs) in enumerate(signals):

    plt.subplot(4, 2, 2*i + 1)
    plt.plot(t, sig)
    plt.title(name)
    plt.xlabel("Czas [s]")
    plt.ylabel("Amplituda")
    plt.xlim(0, 0.005)   # pokaż ~5 ms
    plt.grid(True, alpha=0.3)

    harmonics = np.arange(1, len(coeffs) + 1)
    freqs_h = harmonics * f0

    plt.subplot(4, 2, 2*i + 2)
    plt.stem(freqs_h, coeffs)
    plt.title(f"{name} - składowe harmoniczne")
    plt.xlabel("Częstotliwość [Hz]")
    plt.ylabel("Amplituda")
    plt.xlim(0, f0 * (len(coeffs) + 1))
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### **3. Informatyka**
#### **3.1 Syngały dźwiękowe w formie cyfrowej**
W poprzedniej części pozwoliłem sobie na spore uproszczenie - założyłem, że znamy już skład harmoniczny i na ich podstawie rozwineliśmy funkcję w szereg Fouriera. W praktyce nie dostaniemy tych inforamcji podanych na tacy, a jedynie zdyskretyzowany, cyforwy sygnał. A więc powstaje pytanie, jak z dyskretnych próbek utworzyć szereg Fouriera oraz wydobyć skład harmoniczny?
#### **3.2 Dyskretna transformata Fouriera**
Dyskretna transformata Fouriera pozwala nam przybliżyć faktyczne wartości szeregu lub transformaty Fouriera, kiedy na  przedziale $[0; \pi]$, mamy $N$ równomierne rozmieszczonych punktów. Dany jest wzór:
$$
\Delta t = \frac{2\pi}{N}
$$

$$
t_k = k \Delta t = \frac{2\pi k}{N},  k  = 0, 1, ..., N - 1
$$
$$
f_n \approx \frac{1}{2\pi} \Delta t \sum_{k=0}^{N-1} e^{-int_k} f(t_k)
$$
Podany wyżej wzór jest tak naprawdę przybliżaniej całki metodą prostokątów. W praktyce pozwala nam to pracować na dyskretnych wartościach i przybliżac realne wyniki stosując metrykę $d_2$.

#### 3.3 **Zastosowanie do analizy cyfrowego sygnału**

Mając dyskrteną funkcję dźwiękową o $N$-próbkach, to żeby wyznaczyć ilość składowej o częstotliowści $f_k$, należy wyznaczyć $A_k$

```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np

def DFT(x):
    """
    X[k] = sum_{n=0}^{N-1} x[n] * e^{-j 2π k n / N}
    """
    x = np.asarray(x, dtype=complex)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    W = np.exp(-2j * np.pi * k * n / N)
    return W @ x
try:
    fs, data = wavfile.read("violinC4.wav")

print(f"Częstotliwość próbkowania: {fs} Hz")


x = data.astype(float)
N = len(x)

X = DFT(x)

M = N // 2
X_pos = X[:M]
freqs = np.arange(M) * fs / N

plt.figure(figsize=(12, 5))

Nplot = min(150, N)
plt.subplot(1, 2, 1)
plt.stem(np.arange(Nplot), x[:Nplot], use_line_collection=True)
plt.xlabel("Numer próbki")
plt.ylabel("Amplituda")
plt.title(f"Pierwsze {Nplot} próbek sygnału")

plt.subplot(1, 2, 2)
plt.plot(freqs, np.abs(X_pos))
plt.xlabel("Częstotliwość [Hz]")
plt.ylabel("|X[k]|")
plt.title("Widmo amplitudowe (DFT)")
plt.xlim(0, 4000)

plt.tight_layout()
plt.show()
```

#### **3.4 Rozpoznawanie instrumentów na bazie widma**

Po obliczeniu widma sygnału możemy próbować rozpoznać, jaki instrument zagrał dźwięk. Kluczowa obserwacja jest taka, że każdy instrument 
ma charakterystyczny kształt widma – inną relację między składową podstawową a jej harmonicznymi, inną ilość szumu, inną jasność dźwięku. To właśnie ta barwa różni skrzypce od fletu, nawet jeśli grają tę samą nutę.

Typowy algorytm robi więc w skrócie trzy rzeczy:

1. **Wyciąga cechy z widma** – np. informacje o tym, jak silne są kolejne harmoniczne, ile jest energii w niskich vs wysokich częstotliwościach, jak bardzo „szumiący” jest dźwięk itd.
2. **Tworzy z nich wektor cech** – kilka liczb, które opisują barwę dźwięku w sposób zrozumiały dla komputera.
3. **Porównuje ten wektor z wcześniej nauczonymi wzorcami** dla różnych instrumentów (np. przy pomocy klasyfikatora uczonego maszynowo) i wybiera taki instrument, którego wzorzec jest najbardziej podobny.

W ten sposób, patrząc tylko na widmo i jego ogólny kształt, algorytm jest w stanie oszacować, czy słyszy np. skrzypce, flet, trąbkę czy gitarę.
